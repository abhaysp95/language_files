```python
plot(sin(x ^ 2), -pi, pi, color='green', thickness=2, title='sin($x^2$)')  # title supports latex format
```

```python
plot(sin(x ^ 2), -pi, pi, color='green', thickness=2, legend_label='sin($x^2$)') + plot(cos(x ^ 2), -pi, pi, color='blue', thickness=2, legend_label='cos($x^2$)')
```

```python
p = plot(floor(x), -4, 4, thickness=2, legend_label='floor', exclude=[-3, -2, .., 4])
p.show()
```
